## Obtendo a base de dados

```python
# Para fazer a análise exploratória dos dados
import pandas as pd
# Para gerar os gráficos
import seaborn as sns
# Para customizar o gráficos, ou seja, colocar título, ordenar...
import matplotlib.pyplot as plt

fonte = "https://github.com/isnardsilva/analise-idosos-ENEM2019/blob/master/MICRODADOS_ENEM_2019_IDOSOS.csv?raw=true"
# fonte = "MICRODADOS_ENEM_2019_IDOSOS.csv"

dados = pd.read_csv(fonte)

# Definindo estilo dos gráficos
sns.set_style("darkgrid")
```

```python
dados.columns.values
```

```python
dados.head()
```

## Analisando a idade dos idosos

### Quantidade de idosos por idade

```python
dados['NU_IDADE'].value_counts()
```

```python
dados['NU_IDADE'].value_counts().sort_index()
```

```python
dados['NU_IDADE'].value_counts(normalize = True).sort_index()
```

```python
sns.set(rc={'figure.figsize':(12, 8)})
grafico = sns.countplot(dados.NU_IDADE).set_title("Distribuição dos idosos por idade", fontsize=25)
plt.xlabel("Idade", fontsize=20)
plt.ylabel("Frequência", fontsize=20)
plt.tick_params(labelsize=15)
plt.show()

# fig = grafico.get_figure()
# fig.set_size_inches(12, 8)
# fig.savefig('graficos/distrib_idosos_por_idade.png')
```
